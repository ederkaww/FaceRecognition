```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator 
```

```python
photos_dir = 'photos/photos'

new_gen = ImageDataGenerator(rotation_range=90,
                            horizontal_flip=True,
                            vertical_flip = True,
                            brightness_range=[0.85,1.15],
                            width_shift_range=0.05,
                            height_shift_range=0.05,
                            zoom_range=0.1,
                            fill_mode='nearest')

new_dataset = new_gen.flow_from_directory(photos_dir, 
                                        target_size = (200, 200),
                                        batch_size = 500,
                                        save_to_dir='photos/photos',
                                        save_prefix='new_',
                                        save_format='jpg')

new_dataset.next()
```
